```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
import matplotlib.pyplot as plt
```

### Line Graphs

```python
x1 = [2,3,5,6]
y1 = [1,3,4,5]
x2 = [2,4,6,8]
y2 = [1,2,3,4]

plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
# legends
plt.plot(x1,y1, label='first line')
plt.plot(x2,y2, label='second line')
plt.legend() # only if this line, the labels will be displayed
plt.show()
```

```python
# titles and axis labels
plt.plot(x1,y1, label='first line')
plt.plot(x2,y2, label='second line')
plt.title('Simple Line Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

```python
# line color, width, style
plt.plot(x1,y1, label='first line', color='b', linewidth=5, linestyle='--') # -- for dashed line
plt.plot(x2,y2, label='second line', color='k', linewidth=3, linestyle='-.') # k for black color
plt.title('Simple Line Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

```python
# markers
plt.plot(x1,y1, label='first line', color='b', marker='s', markersize=5) # s for square
plt.plot(x2,y2, label='second line', color='k', marker='o', markersize=5) # o for circle
plt.title('Simple Line Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.show()
```

```python
# grid
plt.plot(x1,y1, label='first line', color='b', marker='s', markersize=5) # s for square
plt.plot(x2,y2, label='second line', color='k', marker='o', markersize=5) # o for circle
plt.title('Simple Line Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid()
plt.show()
```

```python
# sin wave
x = np.arange(0,10,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
# cos wave
x = np.arange(0,10,0.1)
y = np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,10,0.1)
s = np.sin(x)
c = np.cos(x)
plt.plot(x,s, label='Sine Wave')
plt.plot(x,c, label='Cos Wave')
plt.title('Simple Sine and Cosine Wave')
plt.legend()
plt.show()
```

### Bar Graphs

```python
l1 = [5,9,3,11]
l2 = [1,3,5,7]

plt.bar(l2,l1)
plt.show()
```

```python
l3 = [2,4,6,8]
l4 = [9,4,7,5]
plt.bar(l2,l1)
plt.bar(l3,l4)
plt.show()
```

```python
# labels, titles, colors, grids
plt.bar(l2,l1, label='first bar', color='r')
plt.bar(l3,l4, label='second bar', color='g')
plt.title('Simple Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.legend()
plt.grid()
plt.show()
```

```python
df1 = DataFrame({'Revenue':[200,300,350,400,120,600]},
                 index=['Wipro', 'CISCO', 'Google', 'Yahoo', 'Facebook', 'TCS'])
df1
```

```python
plt.bar(df1.index, df1['Revenue'])
plt.show()
```

```python
# horizontal bar graph
plt.barh(df1.index, df1['Revenue'])
plt.show()
```

```python
# rotate labels on x axis, y axis
plt.bar(df1.index, df1['Revenue'])
plt.xticks(rotation=90)
plt.yticks(rotation=45)
plt.show()
```

### Stacked Bars

```python
b1=[12,8,9,10]
b2=[3,10,10,6]
x=[1,2,3,4]
plt.bar(x,b1,label='Male')
plt.bar(x,b2,bottom=b1,label='Female')
plt.legend()
plt.xticks([1,2,3,4],list('ABCD'))
plt.show()
```

### Scatter Plot

```python
x1 = [1,2,3,4]
y1 = [2,4,6,8]
x2 = [2,4,6,8]
y2 = [1,2,3,4]

plt.scatter(x1,y1)
plt.scatter(x2,y2)
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.grid()
plt.show()
```

### Histogram

```python
bsm = [113,85,90,150,149,88,93,115,135,80,77,82,129]
bsf = [67,98,89,120,133,150,84,69,89,79,120,112,100]

plt.hist(bsm) # by default number of bins/intervals is 10
plt.show()
```

```python
# define no of intervals
plt.hist(bsm, bins=4)
plt.show()
```

```python
# shrinking bar width
plt.hist(bsm, bins=4, rwidth=0.95)
plt.show()
```

```python
# define intervals explicitly
plt.hist(bsm, bins=[80,100,125,150], rwidth=0.95)
plt.show()
```

```python
# histogram for two sets of data
# interval will be same for both; it is the width of both bars
plt.hist([bsm,bsf], bins=3, label=['men','women'], color=['k','r'])
plt.show()
```
