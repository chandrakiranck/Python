```python
### During transactions, check if account number exists; if not print error message

ledger = {}
acnum_gen = 0

def createAccount(acname, idnum):
    #Check if ID number already exists    
    for account in ledger.values():
        if account[1] == idnum:
            print("Account with ID Number already exists")
            return
    
    global acnum_gen
    
    # update acnum_gen
    acnum_gen = acnum_gen + 1
    
    print("Creating new Account ({}):".format(acnum_gen))
    
    # Create new account and add to ledger with new account number as key
    newAccount = [acname, idnum, 0, []]
    ledger[acnum_gen] = newAccount    
    return acnum_gen
    
def getBalance(acnum):
    return ledger[acnum][2]

def setBalance(acnum, newBalance):
    ledger[acnum][2] = newBalance
    
def getTransactions(acnum):
    return ledger[acnum][3]

def appendTransaction(acnum, newTransaction):
    transactions = getTransactions(acnum)
    transactions.append(newTransaction)
    
    # if transactions has more than 5 values then remove the first one
    if len(transactions) > 5:
        transactions.pop(0)
    
def doCredit(acnum, amount):
    print("Crediting {} to account number {}".format(amount, acnum))
    setBalance(acnum, getBalance(acnum) + amount)
    appendTransaction(acnum, amount)
    return getBalance(acnum)

def doDebit(acnum, amount):
    print("Debiting {} from account number {}".format(amount, acnum))
    
    if amount > getBalance(acnum):
        print("Insufficient Balance\n")
        return
    
    setBalance(acnum, getBalance(acnum) - amount)
    appendTransaction(acnum, -1 * amount)
    return getBalance(acnum)

def banking():
    print(ledger)
    print()
    
    while True:
        print("enter 1 to 5; anything else to end")
        print("1. Create Account\t2. Credit to Account\t3. Debit from Account\t4. Get Balance\t5. Get Details")
        print()
        
        choice = input()
        if(choice == '1'):
            acname = input("Enter Name: ")
            idnum = input("Enter ID Number: ")
            createAccount(acname, idnum)
        elif(choice == '2'):
            acnum = int(input("Enter Account Number: "))
            amount = int(input("Enter Balance to Credit: "))
            doCredit(acnum, amount)
        elif(choice == '3'):
            acnum = int(input("Enter Account Number: "))
            amount = int(input("Enter Balance to Debit: "))
            doDebit(acnum, amount)
        elif(choice == '4'):
            acnum = int(input("Enter Account Number: "))
            print("Balance =", getBalance(acnum))
        elif(choice == '5'):
            acnum = int(input("Enter Account Number: "))
            print("Transactions =", getTransactions(acnum))
        else:
            print("Exiting...")
            break
        
        print()
        print(ledger)
        print()
            
banking()
```
