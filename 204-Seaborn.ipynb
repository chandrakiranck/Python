## Seaborn

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = sns.load_dataset('tips')
df1.head()
```

```python
df1.tail()
```

```python
sns.get_dataset_names()
```

```python
# used to ensure no warnings pop up
import warnings
warnings.filterwarnings('ignore')

sns.get_dataset_names()
```

### Count Plot

```python
df1['sex'].value_counts()
```

```python
# plot function exists in Pandas as well
df1['sex'].value_counts().plot(kind='bar')
plt.show()
```

```python
# directly get a frequency count on the basis of a column using Seaborn
sns.countplot(df1['sex'])
plt.show()
```

```python
sns.countplot(df1['day'])
plt.show()
```

```python
# further divide bar graph on the basis of another column
# subsections
sns.countplot(data=df1, x='sex', hue='smoker')
plt.show()
```

### Pie Plot

```python
# pie chart using Pandas
df1['day'].value_counts().plot(kind='pie')
plt.show()
```

### Histogram

```python
plt.hist(df1['total_bill'], bins=3, rwidth=0.95)
plt.show()
```

```python
plt.hist(df1['tip'], bins=3, rwidth=0.95)
plt.show()
```

### LM Plot

```python
plt.scatter(df1['total_bill'], df1['tip'])
plt.show()
```

```python
# Linear Modeling Plot
sns.lmplot(data=df1,x='total_bill',y='tip')
plt.show()
```

```python
sns.lmplot(data=df1,x='total_bill',y='tip',fit_reg=False)
plt.show()
```

```python
# divide the graph points on basis of sex
sns.lmplot(data=df1,x='total_bill',y='tip',fit_reg=False,hue='sex')
plt.show()
```

```python
# divide the graphs further on the basis of smokers and time
sns.lmplot(data=df1,x='total_bill',y='tip',fit_reg=False,hue='sex',col='smoker',row='time')
plt.show()
```

### Save Images

```python
# to save images
sns.lmplot(data=df1,x='total_bill',y='tip',fit_reg=False,hue='sex',col='smoker',row='time')
plt.savefig('fig1.png')
```

### Strip Plot

```python
# Strip plot
sns.stripplot(data=df1,x='day',y='total_bill',jitter=False)
plt.show()
```

### Swarm Plot

```python
# Swarm plot does not let points overlap
sns.swarmplot(data=df1,x='day',y='total_bill')
plt.show()
```

```python
sns.swarmplot(data=df1,x='day',y='total_bill')
plt.show()
```

### Joint Plot

```python
# Joint Plot
sns.jointplot(data=df1,x='total_bill',y='tip')
plt.show()
```

### Pair Plot

```python
# Pair Plot
sns.pairplot(df1)
plt.show()
```

### Dist Plot

```python
# Dist Plot
sns.distplot(df1['total_bill'])
plt.show()
```

### Box Plot

```python
# Box Plot
sns.boxplot(df1['total_bill'])
plt.show()
```

```python
sns.boxplot(data=df1,x='day',y='total_bill')
plt.show()
```

### Heat Map

```python
# Heat Map
corrmat = np.random.randint(1,1000,100).reshape(10,10)
sns.heatmap(corrmat)
plt.show()
```

### Factor Plot

```python
# Factor Plot
sns.factorplot('size','total_bill',data=df1,hue='day')
plt.show()
```
